```python
import copy
__params__ =  {'--alias': None,
             '--compile': None,
             '--conf': 'promise.yml',
             '--debug': False,
             '--files': None,
             '--help': False,
             '--log': None,
             '--nbDigits': None,
             '--noParsing': False,
             '--output': None,
             '--path': None,
             '--pause': False,
             '--run': None,
             '--verbosity': '1',
             '--verbosityLog': None,
			 '--CC': 'g++',
			 '--CXX': 'g++',
}


def customArgParser(args, params):
	"Return a user-defined argment list for PROMISE settings."
	params_copy = copy.deepcopy(params)
	args_clean = [i.split('=') for i in args]
	
	for i in args_clean:
		if '--' in i[0]:
			if i[1] in {'True', 'true'}:
				params_copy[i[0]] = True
			elif i[1] in {'False', 'false'}:
				params_copy[i[0]] = False
			else:
				params_copy[i[0]] = i[1]
		else:
			params_copy[i[0]] = True

	return params_copy
```

```python
args = ['--args=2', '--versions=3', 'chsd', '--debug=True']
customArgParser(args, __params__)
```
